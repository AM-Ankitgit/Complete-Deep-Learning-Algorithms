`Non-linearity:` `An activation function introduces non-linear properties to the network,` which enables the model to learn more complex decision boundaries. Most real-world data problems are non-linear, and using an activation function allows the neural network to make sense of such data by learning non-linear hypotheses.

`Control of Neuron Firing:` Neurons in a neural network work in a way somewhat analogous to neurons in a biological brain. The activation function decides whether a neuron should be activated or not, meaning whether the information that the neuron is receiving is relevant for the given information or should be ignored

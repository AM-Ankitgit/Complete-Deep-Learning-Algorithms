# Data Augmentation :
- Creating artifical data from the existing data 



why we need the augementation :
- if we do not have sufficient data
- To reduce the overfitting


```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img(r'D:\study\Complete-Deep-Learning-Algorithms\Data Agumentation CNN\images.jpg',target_size=(200,200))
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
type(img)
```

```python
data_generator = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2)
```

```python
# convert into array 
img = image.img_to_array(img)
img.shape

# 3 because image is RGB (color image)
```

```python
input_batch = img.reshape(1,200,200,3) # how many image what to process here is one image so that's why we need to add this dim
```

```python
i = 0


dir = r'D:\study\Complete-Deep-Learning-Algorithms\Data Agumentation CNN\aug'

# if in file has multiple images want to apply augm on  whole dir use >> flow_from_directory
# for output in data_generator.flow_from_directory(input_batch,batch_size=1,class_mode='binary) # binary classification


for output in data_generator.flow(input_batch,batch_size=1,save_to_dir=dir):
    i = i+1
    if i==10:# create 10 images
        break
```

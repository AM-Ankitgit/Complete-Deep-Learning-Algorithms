```python
def build_model(hp):

    model = keras.Sequential()
    """
    for number of neuron ,type of activation function,
    # unit = hp.Choice('units', [8, 16, 32,6])
    """

    counter = 0

    for i in range(hp.Int('num_layers',min_value=1,max_value=10)):

        number_unit = hp.Int('units'+ str(i), min_value=8,max_value=512,step=8)
        activation_list = hp.Choice('activation' + str(i),values=['relu','tanh','leaky_relu'])
        
        """ 
        Tune Whether to use the dropout layer or not
        in Dropout we drop random neuron from the our input and hidden layer for every epochs
        """

        drop_rate=hp.Float('dropout'+str(i),
                min_value=0.1,
                max_value=0.5,
                step=0.05)
        
        if counter == 0:
            counter+=1

            model.add(Dense(number_unit,activation=activation_list,input_dim=X.shape[1])) # input layer

            # if hp.Boolean('dropout'): # droput layer with condition
            model.add(Dropout(rate=drop_rate))  #after the input layer if the  dropout needed then we can use
        
        else:
            """
            Choose the number of hidden layer 
            """
            
            # total 10 `Hidden layer`
            
            model.add(Dense(number_unit,activation = activation_list))
            
            # if hp.Boolean('dropout'):
            model.add(Dropout(rate=drop_rate))

    """
    Learning Rate using hyperparameter
    Define the optimizer learning rate as a hyperparameter.
    """
    # 0ut side of loop your outout layer
    learning_rate = hp.Float("lr", min_value=1e-4, max_value=1e-2, sampling="log")
    
    model.add(Dense(1, activation='sigmoid')) # ouptput layer
    
    optimizer_list = hp.Choice('optimizer',['adam','sgd','rmsprop','adadelta'])
    
    model.compile(loss='binary_crossentropy',optimizer=optimizer_list,metrics=['accuracy'])
    
    return model
```

```python

'''
Initialize a tuner (here, RandomSearch). We use objective to specify the objective to select the best models, 
and we use max_trials to specify the number of different models to try.
'''

tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy', # what is object here we want less validation loss 
    max_trials=5,
    # directory='ModelHyperparameterSummary',
    # project_name='Diabetes'
    )
# here we can mention directory and project_name to save the summary  model  for each hyperparametrs

```

```python
tuner.search(X_train, y_train, epochs=20, validation_data=(X_test, y_test))


```

```python
best_model = tuner.get_best_models(num_models=1)[0]
```

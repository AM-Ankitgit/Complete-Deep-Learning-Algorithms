# Type of Attention

1. Bahdanau Attention
2. Luong Attention


Without Attention
Context Vector: The encoder processes the input sequence and produces a single fixed-length context vector, typically derived from the final hidden state of the encoder. This single context vector is intended to summarize the entire input sequence and is used to initialize the decoder and guide its generation of the output sequence.
With Attention
Dynamic Context Vectors: Instead of relying on a single fixed-length context vector, the attention mechanism allows the decoder to create a dynamic context vector at each step of the decoding process. This context vector is a weighted sum of all the encoder's hidden states, where the weights are computed based on the decoder's current state and the encoder's hidden states.
How Attention Works with Context Vectors
Compute Attention Scores: At each decoding step, the decoder computes a set of attention scores for each of the encoder's hidden states. These scores indicate how much attention should be paid to each hidden state of the encoder.
Attention Weights: The attention scores are normalized using a softmax function to produce attention weights.
Dynamic Context Vector: The dynamic context vector is then computed as a weighted sum of the encoder's hidden states, using the attention weights.
This dynamic context vector is updated at each decoding step, allowing the decoder to focus on different parts of the input sequence as needed.

Summary
Fixed Context Vector: Used in traditional seq2seq models without attention. A single context vector summarizes the entire input sequence.
Dynamic Context Vector: Used in seq2seq models with attention. The context vector is dynamically computed at each decoding step, allowing the model to focus on different parts of the input sequence as needed.

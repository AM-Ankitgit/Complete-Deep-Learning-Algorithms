## Architecture of RNN and LSM

here RNN maintaining single state and the LSTM maintaining two state long and short term memory state

Forget Gate :
- it decide the how much information should have to forget from the long term memory based on the current input

Input Gate :
- it decide to the how much new  information has to be added to the memory cell (long term memory)


Output Gate :
- based on the current input it decide the which output has to give as from the long term memory

LTM : long term memory called as the cell state (Ct)

STM : short term memory called as the hidden layer(ht)

and it also has the communication between the LTM and STM

here forgate gate something from the current cell state 

and the input gate adding something new to cell state

 this process is called updating the cell state (mean LTM >>>>Ct)

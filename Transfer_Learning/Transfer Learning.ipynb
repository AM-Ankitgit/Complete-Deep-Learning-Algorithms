## `What is Transfer Learning`



Transfer learning, used in machine learning, is the reuse of a pre-trained model on a new problem.

- In transfer learning we use the pre-trained model, 
from pre-trained model we take the convolution base and we drop the FC connected layer of pre-trained model(make convolution base freeze so weight values should not be change)
and add our own FC connected layers (So here we create our own  FC layer arch as per needs)

`Why this method pre- trained model work:`
it all ready learn the general details (primitive features or simple feature ,like edges) of the images 


`Feature Extraction` : We use the convolution base as it is and we train the FC connected layer<br>
`Fine Tunning `: In fine tunning we  train the few last layer of convolution base and all layer of FC

`when to use`:


`Feature Extraction` : if your problem of classification is matching to your pre-trained model data set

`Fine Tunning `: if your problem of classification is not matching to your pre-trained model data set(like tablet and phone)

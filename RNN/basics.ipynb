# Architecture of RNN

- RNN and ANN has two big difference in Arch

`Data feeding to NN`:
- ANN whole data we can pass without timestamps
- RNN data feed based on timestamps(t1 -->X1 input and for tn --->Xn)
- ANN are feed Farward (inforamtion move from input to output side)
- RNN can hold state of output and give feedback to input of each neuron or node










for first time stamp for t1 feedback will be random or may be 0

# arch in code

```python
from keras import Sequential
from keras.layers import Dense,SimpleRNN

# in RNN by default activation is tanh

model = Sequential()

model.add(SimpleRNN(3,input_shape=(4,5))) # where 4 = timestep and 5 = input feature
model.add(Dense(1,activation='sigmoid'))

model.summary()
```
